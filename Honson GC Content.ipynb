```python
#Letters not permitted in sequence
letters = ["B","D","E","F","H","I","J","K","L","M","N","O","P","Q","R","S","U","V","W","X","Y","Z"]
#Sequence Input
seq = input("Hello, what is your sequence?   ")
sequ = seq.upper()
#Error message if non-base letters
if any(elem in letters for elem in sequ):
    print("I'm sorry, there are non-base letters in your sequence")
#Removes spaces from sequence, choose which bases to count
else:
    import re
    seqf = re.sub(" ", "", sequ)
    ct = input("Okay, what would you like to count?   ")
#Convert all letters to uppercase 
    ctu = ct.upper()
#Search for characters of interest, assign uncalled characters to zero
    if "A" in ctu:
        a = seqf.count("A")
    else: 
        a = 0
    if "T" in ctu:
        t = seqf.count("T")
    else:
        t = 0
    if "C" in ctu:
        c = seqf.count("C")
    else:
        c = 0
    if "G" in ctu:
        g = seqf.count("G")
    else:
        g = 0
#Sum character values. Unasked for characters will be zero.
    total = a + t + c + g
#Calculate and print percents
    percent = total/len(seqf)
    print(ctu + " Count: " + str(total))
    print(ctu + " Content: " + str(round(percent,3)))
```
